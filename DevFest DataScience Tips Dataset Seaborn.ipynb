```python
import seaborn as sns 
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
sns. relplot(data=df,x="total_bill", y="tip")
#like a scatter plot
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker", style="smoker")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker", style="smoker", size="size")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",col="time")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",col="time",row="size")
#eatind dinner or lunch alone
```
